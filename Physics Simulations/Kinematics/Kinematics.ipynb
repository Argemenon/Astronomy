# Kinematics

### Problem 1

+ A cart is on a track at x=0 at t=0
+ The cart is moving with velocity 0.45 m/s
+ Where is the cart at t=1.5 sec

```python
import vpython
import matplotlib.pyplot as plt
```

```python
x = 0
t = 0
v = 0.45
dt = 0.25

while t < 1.5:
    x = x + v*dt
    t = t + dt
    print("x = ", x)
```

### Problem 3

+ A cart starts at x = 0, t = 0 with velocity of 0.45
+ Cart has an acceleration of -1.4
+ where is the cart after 15 seconds?
+ pos time graph?

```python
x = 0
t = 0
dt = 0.01
v = 10
a = -1.4

pos = list()
vel = list()
tim = list()

while t < 15:
    x = x + v*dt
    pos.append(x)
    v = v + a*dt
    vel.append(v)
    t = t + dt
    tim.append(t)
print(x, v)
```

```python
plt.plot(tim, pos)
```

```python
plt.plot(tim, vel)
```

```python
x = 0
t = 0
dt = 0.01
v = 10
a = -1.4

pos = list()
vel = list()
tim = list()

while v >= 0:
    x = x + v*dt
    pos.append(x)
    v = v + a*dt
    vel.append(v)
    t = t + dt
    tim.append(t)
print(x, v)
```
